```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# input dataset
df= '/content/StudentPerformanceFactors.csv'
data = pd.read_csv(df)

# memilih kolom yang relevan untu dilakukan analisis
relevant_columns = ['Hours_Studied', 'Attendance', 'Motivation_Level', 'Exam_Score']

# mengubah Motivation_Level menjadi numerik
data['Motivation_Level_Numerical'] = data['Motivation_Level'].map({'Low': 1, 'Medium': 2, 'High': 3})
relevant_columns = ['Hours_Studied', 'Attendance', 'Motivation_Level_Numerical', 'Exam_Score']

# Menghitung matrix korelasi 
corr_matrix = data[relevant_columns].corr()

# Visualisai Heatmap untuk korelasi 
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation between Factors and Exam Score')
plt.show()

# Scatter plot antara Hours Studied and Exam Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=data, hue='Motivation_Level', palette='Set2') # You can still use the original 'Motivation_Level' for hue
plt.title('Hours Studied vs Exam Score by Motivation Level')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```
